```python
def num2color(values,cmap,vmin = None):
    cmap = plt.get_cmap(cmap)
    norm = plt.Normalize(vmin=vmin,
                         vmax=np.max(values))
    res = []
    for v in values:
        res.append(cmap(norm(v)))
    return res
```

```python
pm_hour = inhale_data1.groupby('hour_of_day').mean()
plot_title = "PM Averages per day by hour"
colors = num2color(pm_hour["pm2_5"],'OrRd',0)
pm_hour['pm2_5'].plot.bar(figsize=(10,6),color=colors)
plt.title("Average PM2.5 exposures by hour of day")
plt.xlabel("Hour of the day")
plt.ylabel("PM level (ug/m3)")
plt.xticks(rotation=45)
plt.tight_layout()
# plt.savefig(f"../new_london_data_eda/plot/{plot_title}.png", dpi=300)
```
